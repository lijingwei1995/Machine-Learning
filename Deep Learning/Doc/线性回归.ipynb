```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# 生成4个点
np.random.seed(2)

point_num  = 10
scope = 10

X = np.array([np.random.randint(-scope, scope) for _ in range(point_num)])
Y = np.array([np.random.randint(-scope, scope) for _ in range(point_num)])


print(X, Y)
```

```python
# 绘制
fig = plt.figure()
plt.plot(X, Y, "o")
plt.show()
```

```python
# 画一张损失函数关于(a,b)的图像
a = np.arange(-0.23, -0.21, 0.0001)
b = np.arange(-2.2, 1.0, 0.01)
c = np.ones(point_num)

a, b, c = np.meshgrid(a, b, c)

# shape(a) & shape(b) = (200, 200, 4)

X_ = X.reshape(1, 1, -1)
Y_ = Y.reshape(1, 1, -1)

# shape(X) & shape(Y) = (1, 1, 4)

# trailing dimension broadcasting

Loss = (Y_ - a * X_ + b) ** 2
# print(Loss)
# shape(Loss) = (200, 200, 4)

Loss_sum = Loss.sum(axis = 2)
# print(Loss_sum)
# shape(Loss_sum) = (200, 200)

fig = plt.figure()
ax = Axes3D(fig)

ax.plot_surface(a[:,:,0], b[:,:,0], Loss_sum)

plt.show()
```

```python
# 学习率
alpha = 0.0001

# 设 y = ax + b
# 随机初始化
a = 2
b = -2

# 收敛阈值
threhold = 0.1
```

```python
# 批量梯度下降 BGD
iteration = 1
batch = point_num
Loss_sum = float("inf")
while Loss_sum > threhold and iteration < 1000:
    for i in range(0, point_num, batch):
        d_a, d_b = 0, 0
        for j in range(i, min(i + batch, point_num)):
            d_a += alpha * (Y[j] - a * X[j] - b) * (-X[j])
            d_b += alpha * (Y[j] - a * X[j] - b) * (-1)
        print(d_b)
        a -= d_a
        b -= d_b
    Loss = (Y - a * X + b) ** 2
    Loss_sum = Loss.sum()
    print("iteration:%d, a:%f, b:%f, Loss:%f" % (iteration, a, b, Loss_sum))
    iteration += 1
```
